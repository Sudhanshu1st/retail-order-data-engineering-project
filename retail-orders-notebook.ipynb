### Reading the file 

```python
import pandas as pd 
df = pd.read_csv(r"C:\Users\sudha\My Projects\retail-order-sql+python-project\orders.csv", na_values = ['Not Available', 'unknown'])

df.head(5)
```

```python
df['Ship Mode'].unique()
```

```python
#rename columns names and replace space with underscore

#df.rename( columns ={'Order Id': 'order_id', 'City': 'city'})
```

```python
df.columns = df.columns.str.lower()
```

```python
df.columns = df.columns.str.replace(' ', '_')
```

```python
df.head()
```

```python
#Let's make new columns discount, sales price, profit 

df['discount'] = df['list_price']*df['discount_percent']*0.01
```

```python
df['sale_price'] = df['list_price'] - df['discount']
```

```python
df['profit'] = df['sale_price'] - df['cost_price']
```

```python
#checking datatype 

df.dtypes
```

```python
#converting orderdate datatype to date datatype

pd.to_datetime(df['order_date'], format= "%Y-%m-%d")
```

```python
#dropping unnecessary columns 

df.drop(columns = ['list_price','cost_price', 'discount_percent'], inplace =True)
```
